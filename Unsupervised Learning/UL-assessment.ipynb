**Question 5** (12 marks)

Visualize the three fitted mixture distributions (side-by-side in a 1 x 3 split window) 


```r
# YOUR CODE HERE
plot.mix<- function(dat, p, mu, sigma, breaks=26, dens=TRUE, ngrid=401, ...){
  try<-  hist(dat, breaks=breaks,  plot=FALSE)
  hist(dat, breaks=breaks, freq=FALSE, ylim=c(0, max(try$density)*1.3), col="grey93" , border="grey85",...)
  r <- diff(range(dat))
  grid<- seq(min(dat)-0.15*r, max(dat)+0.15*r, length=ngrid)
  K<- length(p)
  if (length(sigma)==1){
    sigma<-rep(sigma, K)
  }
  grid.mat<- matrix(0, ngrid,K)
  for (j in 1:K){
    grid.mat[,j]<- p[j]*dnorm(grid, mu[j], sigma[j])
  }
  for (j in 1:K){
    lines(grid, grid.mat[,j], col=j+1, lwd=2)
  }
  if (dens){
    lines(grid, apply(grid.mat,1, sum), col=1,lwd=2)
  }   
  invisible()  
}


### 1 x 3 split window but not pretty
par(mfrow=c(1,3),widths=c(5,5))
plot.mix(stamps, mix2$lambda, mix2$mu, mix2$sigma, main="Stamp mixture distributions\n with k = 2 ", xlab="Thickness of Stamps") 
plot.mix(stamps, mix3$lambda, mix3$mu, mix3$sigma, main="Stamp mixture distributions\n with k = 3 ", xlab="Thickness of Stamps")
plot.mix(stamps, mix4$lambda, mix4$mu, mix4$sigma, main="Stamp mixture distributions\n with k = 4 ", xlab="Thickness of Stamps")

########################################################
#Also plotted the graphs individually for better visual#
# was told to do this if I thought it added value      #
########################################################
par(mfrow=c(1,1))

plot.mix(stamps, mix2$lambda, mix2$mu, mix2$sigma, main="Stamp mixture distributions with k = 2 ", xlab="Thickness of Stamps") 
par(mfrow=c(1,1))
plot.mix(stamps, mix3$lambda, mix3$mu, mix3$sigma, main="Stamp mixture distributions with k = 3 ", xlab="Thickness of Stamps")
par(mfrow=c(1,1))
plot.mix(stamps, mix4$lambda, mix4$mu, mix4$sigma, main="Stamp mixture distributions with k = 4 ", xlab="Thickness of Stamps")
```

For the stamps data, carry out a bootstrap likelihood ratio test of H_0: K=2 versus H_1: K=3.  

You will need for this, initially, a function which simulates data from a given mixture model. 
You can use for this purpose any functions that were developed in lectures or labs, or that you find within R.
The bootstrap routine as such MUST be manually implemented.

```r
# YOUR CODE HERE
set.seed(20) 
mix2 <- normalmixEM(stamps, k=2)
mix3 <- normalmixEM(stamps, k=3)

gauss.mix.sim<-function(n, p, mu, sigma){          
   x  <- runif(n)
   sim <-rep(0,n)
   cp <-cumsum(p)
   for (i in 1:n){
       k <-1
       while (x[i]>cp[k]){
           k<-k+1
       }    
       sim[i] <- rnorm(1,mu[k],sigma[k])
   }
   return(sim)
 }

####### STEP 1 ######

LRTS <- -2*mix2$loglik + 2*mix3$loglik

####### STEP 2 ######

M <- 99
sim.LRTS <-rep(0,M)
n <- length(stamps)

for (m in 1:M){
    sim_data <- gauss.mix.sim(n, p=mix2$lambda, mu=mix2$mu, sigma=mix2$sigma)
    sim2<-  normalmixEM(sim_data, k=2, verb=FALSE)
    sim3<-  normalmixEM(sim_data, k=3, verb=FALSE)
    sim.LRTS[m]<- -2*sim2$loglik + 2*sim3$loglik
}

####### STEP 3 ######
pval <- 1-(rank(c(LRTS, sim.LRTS))[1]/100)
pval


###############################
#          COMMENTS           #
###############################
"The p value is greater than 5% therefore we do not reject 𝐻0.
The reason for this is that the OBSERVED LR is NOT among the top 5% 
of the B bootstrapped LR values. This means the observed value we found
is NOT very unlikely under the null hypothesis. This means we can belive 
that it is more likely for our Gaussian mixture model to have K=2 components than 
k = 3 components."
```

**Question 8** (6 marks)

We return now to the problem of kernel density estimation. 

A popular rule for automatic bandwidth selection is Silverman's rule of thumb, which is given by

$$ h = 0.9 \times A n^{(-1/5)}$$

where $A= min(s, IQR/1.34)$, with $s$ being the sample standard deviation of the data, $IQR$ the interquartile range, and $n$ the sample size.

Produce a function with name `hsil` which implements this rule. Then apply it to the stamps data set, and save the resulting bandwidth to an object `h2`.


```r
# YOUR CODE HERE
hsil <- function(x){          
    n <- length(x)
    
    A <- min( sd(x), (IQR(x)/1.34))

    h <- 0.9*A*n^(-1/5)
    
   return(h)
 }
h2 <- hsil(stamps)
round(h2,4)
```

This normal reference allowed us to find a simple expression for the bandwidth by making the assumption that the kernel/function to be normal. This is helpful if our data is normal/binormal however we do see oversmoothing with non normal data. However with alot of data following normal distributions this fourmula is useful in application. What are clusters?

Clusters in a set of data are 'ideally' homogenous sub groups within that dataset. This makes sense intuitively as one expects members of the same cluster/group to be similar to one another. You can think of clusters as data rich regions of the data set which are usually surrounded by some data poor regions. 

what are density modes?

Density/density function provides us a way of understanding what is likely to be observed from a sample data set. The higher the density/value from the function can be interpreted as being more likely. The mode is simply the value from a dataset that appears most often.

So clusters are a homogenouse sub group of data, and as stated before they are data rich or in other words high in density. Therefore by finding the mode(s) of the density distribution of the data set you have a point in which it is likely for a random variable to end up at or near by, you can think of this as how the clusters came about. E.g if we have two clustering in a data set, we have two high density regions. By finding the two highest modes we can identfy the two clusters and more specifically we can say these modes describe on average all mebers within that cluster. If we were to sample a random variable from this data set we can imagine it belonging to one of these two clusters, i.e due to the high density regions around these cluster is is more likely a random sample to be identied as a member of one of these clusters.
