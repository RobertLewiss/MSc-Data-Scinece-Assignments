```r
t <- seq(pi/2+0.2,3*pi/2-0.2, 
    length=60) # Parametriziation
cx <- cos(t); cy <-sin(t) 
      # gives a circular arg segment
cx2 <- cx + rnorm(60,0,0.1); cy2 <- cy + rnorm(60,0,0.1)
      # adds noise
c2 <- cbind(cx2,cy2)
plot(c2)
```

```r
require(princurve)
C.hs <- principal_curve(c2, smoother="lowess")
plot(c2)
lines(C.hs, lwd=2, col=3)
lines(cbind(cx,cy), col=2)
legend(-1.15, 1.1,  c("true C", "principal curve"),  lwd=c(2,1), col=c(3,2))
```

#### Bonus example (not part of lecture slides)

Measurements of average speed and flow values over 444 5-minute-intervals, between 9th of July 2007, 9am, to 10th of July 2007, 10pm, on a Californian Freeway, with principal curve.

```r
data(calspeedflow)
speedflow.curve  <- lpc(calspeedflow[,3:4])
plot(speedflow.curve)
```

We add the orthogonal projections onto a B-spline approximation of the curve:

```r
plot(speedflow.curve, type=c("spline", "project"))
```

Now we can relate the parametrization to physical properties of the system; e.g. the traffic density (i.e., flow/speed) which monotonically increases from the top left (almost no traffic) to the bottom right of the above curve (congestion):

```r
speedflow.project <- lpc.project(speedflow.curve, newdata=calspeedflow[,3:4])
plot(speedflow.project$closest.pi, calspeedflow$Lane5Flow/calspeedflow$Lane5Speed, ylab="traffic density", xlab="curve parametrization")
```

Note that, what you see here, may be either an increasing or a decreasing relationship, depending on the "end" of the curve where the parametrization has been initialized. This is an arbitrary decision.

This curve could be the object of further analysis, but we stop here.
